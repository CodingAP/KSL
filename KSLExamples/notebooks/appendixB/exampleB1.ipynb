```kotlin
val rand = NormalRV()
val data = mapOf(
    "rating" to List(200) { rand.value } + List(200) { rand.value * 1.5 + 1.5 },
    "cond" to List(200) { "A" } + List(200) { "B" }
)

var p = letsPlot(data)
p += geomDensity(color = "dark_green", alpha = .3) { x = "rating"; fill = "cond" }
p + ggsize(700, 350)
```
