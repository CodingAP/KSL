```kotlin
import ksl.utilities.statistic.*
```

```kotlin
val f = IntegerFrequency()
for(i in data){
    f.collect(i as Int)
}
f.toString()
```

```kotlin
val fp = f.frequencyPlot()
```

```kotlin
val plot  = fp.buildPlot()
```
